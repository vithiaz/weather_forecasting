```python
ATTR_NOW = 'sunshine'
```

```python
attr_dict = {
    'min_temp': ('mnt', 'Minimum Temperature', u"\u2103"),
    'max_temp': ('mxt', 'Maximum Temperature', u"\u2103"),
    'avg_temp': ('avgt', 'Average Temperature', u"\u2103"),
    'avg_humidity': ('hum', 'Humidity', "Percentage (%)"),
    'sunshine': ('sun', 'Sunshine', "Hours"),
    'rainfall': ('rain', 'Rainfall', "milimeter (mm)"),
}

ATTR_CODE = attr_dict[ATTR_NOW][0]
ATTR_NAME = attr_dict[ATTR_NOW][1]
ATTR_UNIT = attr_dict[ATTR_NOW][2]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(ATTR_NOW+'.csv')
df.date = pd.to_datetime(df.date, format='%Y-%m-%d')
df.set_index('date', inplace=True)
df.head()
```

```python
def plot_forecasts_result(res_df):
    title = ATTR_NAME + ' Forecasting Results with ARIMA Model'
    
    plt.figure(figsize=(18,8))
    plt.title(title)
    plt.xlabel("Periode", fontsize=12)
    plt.ylabel(ATTR_UNIT, fontsize=12)
    plt.plot(res_df['scn-1'], 'r-', label='1-step forecast')
    plt.plot(res_df['scn-2'], 'b-', label='3-step forecast')
    plt.plot(res_df['scn-3'], 'g-', label='7-step forecast')
    plt.plot(res_df['Actuals'], color='black', linestyle=':' , label='Actuals')
    plt.legend()
```

```python
plot_forecasts_result(df)
```
