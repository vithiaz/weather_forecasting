```python
df = pd.read_csv('../attributes_dataset/max_temp.csv')
df.date = pd.to_datetime(df.date, format='%Y-%m-%d')
df.set_index('date', inplace=True)
df.head()
```

# Splitting Data

```python
train_ratio = int(len(df) * 0.9)
train_ratio
```

```python
dtrain = df.iloc[:train_ratio]
dtest = df.iloc[train_ratio:]
```

# Normality Test

```python
import scipy
from scipy import stats
```

```python
dtrain.max_temp.values
```

```python
stats.probplot(dtrain.max_temp.values, plot=plt)
plt.figure()
```

```python
temp = stats.shapiro(dtrain.max_temp.values)
temp
```

```python
# p > 0.05, data is normal
temp[1] > 0.05
```

# Visualisasi ACF dan PACF

```python
fig = sgt.plot_acf(dtrain, lags=180, zero=False)
fig.set_figheight(8)
fig.set_figwidth(18)
```

Grafik ACF menurun secara eksponensial, ini berarti data sudah stasioner

jika belum grafik menurun secara perlahan, data belum stasioner sehingga perlu dilakukan differencing..

untuk lebih membuktikan bahwa data belum stasioner, maka dilakukan ACF test

```python
pacf_fig = sgt.plot_pacf(dtrain, lags=180, zero=False)
pacf_fig.set_figheight(8)
pacf_fig.set_figwidth(18)
```

```python
comp = seasonal_decompose(dtrain, model='additive')
plt.figure(figsize=(14,4))
comp.seasonal.plot()
```

# Open the Model

```python
## open the model

with open('temp_arima_mxt.pkl', 'rb') as pkl:
    pickle_preds = pickle.load(pkl)
```

```python
pickle_preds.summary()
```

```python
temp_predictions = pd.DataFrame(pickle_preds.predict(n_periods = len(dtest),
                                              exogenous=test_exg),
                            index = dtest.index)
```

```python
plt.figure(figsize=(18,8))
plt.plot(temp_predictions, color='red')
plt.plot(dtest)
```

```python
test_MAE = mean_absolute_error(temp_predictions, dtest)
test_RMSE = sqrt(mean_squared_error(temp_predictions, dtest))

print(f'Mean Absolute Error is\t: {test_MAE}')
print(f'Root Mean Squared Error is\t: {test_RMSE}')
```

# Update

```python
mod_up = pickle_preds.update(dtest, exogenous=test_exg)
```

```python
mod_up.summary()
```

```python
temp_predictions_up = pd.DataFrame(mod_up.predict(n_periods = len(dtest),
                                              exogenous=test_exg),
                            index = dtest.index)
```

```python
plt.figure(figsize=(18,8))
plt.plot(temp_predictions_up, color='blue')
plt.plot(temp_predictions, color='yellow')
plt.plot(dtest, color='green')
```
