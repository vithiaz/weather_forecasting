```python
NEURON_LAYER = 256
EPOCH = 250
ATTR_NOW = 'rainfall'
```

```python
attr_dict = {
    'min_temp': ('mnt', 'Minimum Temperature', u"\u2103"),
    'max_temp': ('mxt', 'Maximum Temperature', u"\u2103"),
    'avg_temp': ('avgt', 'Average Temperature', u"\u2103"),
    'avg_humidity': ('hum', 'Humidity', "Percentage (%)"),
    'sunshine': ('sun', 'Sunshine', "Hours"),
    'rainfall': ('rain', 'Rainfall', "milimeter (mm)"),
}

ATTR_CODE = attr_dict[ATTR_NOW][0]
ATTR_NAME = attr_dict[ATTR_NOW][1]
ATTR_UNIT = attr_dict[ATTR_NOW][2]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df_call = pd.read_csv(ATTR_CODE+'_n'+str(NEURON_LAYER)+'_e'+str(EPOCH)+'.csv')
df_call.date = pd.to_datetime(df_call.date, format='%Y-%m-%d')
df_call.set_index('date', inplace=True)
df_call.head()
```

```python
def plot_forecasts_result(res_df):
    title = ATTR_NAME + ' Forecasting Results with LSTM Model'
    
    plt.figure(figsize=(18,8))
    plt.title(title)
    plt.xlabel("Periode", fontsize=12)
    plt.ylabel(ATTR_UNIT, fontsize=12)
    plt.plot(res_df['scn-1'], 'r-', label='1-step forecast')
    plt.plot(res_df['scn-2'], 'b-', label='3-step forecast')
    plt.plot(res_df['scn-3'], 'g-', label='7-step forecast')
    plt.plot(res_df['Actuals'], color='black', linestyle=':' , label='Actuals')
    plt.legend()
```

```python
def col_forcasts_err(res_df):
    for col in [x for x in range(len(res_df.columns)-1)]:
        col_name = res_df.columns[col]
        MAE = mean_absolute_error(res_df['Actuals'], res_df[col_name])
        RMSE = sqrt(mean_squared_error(res_df['Actuals'], res_df[col_name]))
        print('Error for ', col_name)
        print(f'Mean Absolute Error is\t\t: {MAE}')
        print(f'Root Mean Squared Error is\t: {RMSE}')
        print()
```

```python
plot_forecasts_result(df_call)
```

```python
col_forcasts_err(df_call)
```
